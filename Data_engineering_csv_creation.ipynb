```python
import json
from tqdm import tqdm
def json_readr(file):
    for line in open(file, mode="r", encoding = 'utf-8'):
        yield json.loads(line)
```

```python
from tqdm import tqdm
fields = set()
fields_count = dict()
ref_count = dict()
citations_count = dict()
data = json_readr('dblp_papers_v11.txt')
for obj in tqdm(data):
    if 'fos' in obj:
        for i in obj['fos']:
            if 'name' in i:
                if i['name'] in fields:
                    fields_count[i['name']] += 1
                    if 'references' in obj:
                        ref_count[i['name']] += len(obj['references'])
                    if 'n_citation' in obj:
                        citations_count[i['name']] +=obj['n_citation']
                else:
                    fields.add(i['name'])
                    fields_count[i['name']] = 1
                    if 'references' in obj:
                        ref_count[i['name']] = len(obj['references'])
                    else:
                        ref_count[i['name']] = 0
                    if 'n_citation' in obj:
                        citations_count[i['name']] =obj['n_citation']
                    else:
                        citations_count[i['name']] = 0
```

```python
import operator
sorted(fields_count.items(), key =operator.itemgetter(1), reverse = True)
```

```python
# Creating the nodes file
import csv
authors_database = set()
data = json_readr('file.txt')
with open('nodes_correct.csv', 'w', encoding="utf-8") as csvFile:
    writer = csv.writer(csvFile)
    #writer.writerow(['id', 'name', 'organization'])
    for obj in tqdm(data):
        for i in obj['authors']:
            if i['id'] not in authors_database:
                authors_database.add(i['id'])
                if 'org' in i:
                    writer.writerow([i['id'],'Authors name', i['name'], 'Authors affiliation', i['org']])
                else:
                    writer.writerow([i['id'],'Authors name', i['name'],'Authors affiliation', ])
csvFile.close()
```

```python
# Creating set of papers
data = json_readr('file.txt')
papers = set()
for obj in tqdm(data):
    papers.add(obj['id'])
```

```python
# Creating disctionary of authors for each paper
data = json_readr('file.txt')
paper_authors = {key: [] for key in papers}
for obj in tqdm(data):
    for i in obj['authors']:
        paper_authors[obj['id']].append(i['id'])
```

```python
# creating edges file
# 0 - colaboration, 1 - reference
# For references two papers id: the main one, from where reference is done
# and the referenced one - to which the reference is done
colab = 0
refer = 0
count = 0
data = json_readr('file.txt')
with open('edges_correct.csv', 'w', encoding="utf-8") as csvFile:
    edges = csv.writer(csvFile)
    #edges.writerow(['id_out', 'id_in', 'edge_type', 'paper_main', 'paper_referenced'])
    for obj in tqdm(data):
        for i,author in enumerate(obj['authors']):
            for j in range(i+1,len(obj['authors'])):
                edges.writerow([count, author['id'], obj['authors'][j]['id'], 'Edge label', 0,
                                'main_paper', obj['id'],'cited_paper',])
                count += 1
                colab += 1
            if 'references' in obj:
                for j in obj['references']:
                    if j in papers:
                        for paper_auth in paper_authors[j]:
                            edges.writerow([count, author['id'], paper_auth, 'Edge label',1, 
                                            'main_paper', obj['id'], 'cited_paper', j])
                            refer += 1
                            count+=1
csvFile.close()
```
